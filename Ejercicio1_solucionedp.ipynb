# Ejercicio 1. Grafica de la solución 

La solución de la ecuación 
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2}$$
sujeta a las condiciones de frontera:
$$u(0,t)=u(1,t)=0, \;\;\; t>0$$
Y la condición inicial:
$$u(x,0)=2x+1,\;\;\;0<x<1$$
Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,t)=\sum_{n=1}^{\infty}[1-3(-1)^n]\left(\frac{2}{nπ}\right)sen(n\pi x)e^{-t(nπ)^2}.$$

Este problema modela la transferencia de calor en una barra ideal delgada, con $k=1$.

```python
from pylab import *
```

Graficamos la condicion inicial:

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 2*x[i]+1
    return y
```

```python
x = linspace(0,5,1000)
f = recta(x)
plot(x,f,label='$u(x,0)=2x+1$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-0.01,2.1]),ylim([-0.01,5.01]),grid(),legend()
show()
```

Para aproximar la condición inicial, usamos una serie de Fourier $u(x,0)=f(x)=2x+1$.

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (1.0-3.0*(-1.0)**n)*(2.0/(n*pi))*sin(n*pi*x)
    return f
```

```python
f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=2x+1$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,3.5]),grid(),legend()
show()
```

Ahora veamos que podemos visualizar la aproximacion para diferentes cantidades de armónicos.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i) 
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,3.5]),grid(),legend()
show()
```

Agregamos la parte temporal para visualizar la solución completa.


Para el momento inicial, $t=0$ tenemos:

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += (1.0-3.0*(-1.0)**n)*(2.0/(n*pi))*sin(n*pi*x)*exp(-n**2*pi**2*t[i])
    return (2.0/pi)*f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
```

```python
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.01,2.5]),grid(),legend()
show()
```

Ahora para diferentes tiempos, lo cual nos permite ver cómo la solución evoluciona cuando $t \to \infty$.

```python
plot(x,U[:,0],label='$u(x,t=0)$') 
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$'); 
plt.title("Evolución de la solución")
xlim([-0.01,1.01]),ylim([-0.01,2.5]),grid(),legend()
show()
```

De igual forma podemos verlo en una gráfica 3D

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
t = linspace(0,1.0,100)

X,T = np.meshgrid(x,t)

U = u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$'); 
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="viridis")
```
